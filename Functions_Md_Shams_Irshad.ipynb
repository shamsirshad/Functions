**6. Explain the concept of generators in Python and how they are defined ?**

Ans:    A generator is a special type of iterator in Python that simplifies the process of creating iterators. Generators allow you to yield items one at a time, rather than returning them all at once, which is memory-efficient for large datasets or streams of data.

1. Generators are defined using a function with the yield keyword.
2. When a generator function is called, it doesn’t execute immediately. Instead, it returns a generator object that can be iterated over.
3. Each time yield is encountered, the function's state is saved, and the value is returned to the caller.
4. Execution resumes from the last yield statement when next() is called on the generator.

**"Generators are defined like regular functions but use yield instead of return."**

eg.   
def simple_generator():   
    yield 1   
    yield 2   
    yield 3   

gen = simple_generator()

print(next(gen))  # Output: 1   
print(next(gen))  # Output: 2   
print(next(gen))  # Output: 3   
##### print(next(gen))  # Raises StopIteration because the generator is exhausted

**7.What are the advantages of using generators over regular functions ?**

Ans :   
**Memory Usage :**	Regular function allocates memory for all results at once.	while generator function produces items one at a time, conserving memory.    
**Performance	:** Regular function slower for large datasets due to memory overhead, while generator function is faster due to lazy evaluation.   
**Use Case	:** Regular function best for small datasets or when all results are needed.while generator function best for large datasets or when partial results are needed.   
**Implementation :**	regular function returns all results at once with return, while generator function uses yield to produce values incrementally.

**Example : Using a Regular Function:**   
def squares_regular(n):   
    result = [ ]    
    for i in range(1, n + 1):      
        result.append(i ** 2)   
    return result

##### **Calling the regular function**
n = 5   
squares = squares_regular(n)   
for square in squares:   
    print(square)   


####Output Sequencely :   
1,4,9,16,25

**Issues with Regular Function:**
 * All the results are stored in memory at once, which can be inefficient for large "n".
 * The entire list must be created even if we only need a few values.      


               
**Example : Using a generator Function:**    
def squares_generator(n):   
    for i in range(1, n + 1):   
        yield i ** 2   

##### Calling the generator function
n = 5
for square in squares_generator(n):   
    print(square)   
####Output Sequencely :   
1,4,9,16,25


**Advantages of the Generator:**
* Memory Efficiency: Values are produced one at a time, without storing the entire list in memory.
* Lazy Evaluation: If you stop iterating early, unused values are never computed.

**9.Explain the purpose and usage of the `map()` function in Python ?**

**Purpose :**

* To simplify applying a function to each element of an iterable.
* To make code more concise and readable by replacing explicit loops.

**Example 3:**

**10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?**

Ans :   The map(), reduce(), and filter() functions in Python are higher-order functions used for processing iterables in a functional programming style.

**map() Function :** Purpose: Applies a function to every item in an iterable and returns an iterator containing the results.Transforms the input iterable.Does not filter or combine items.

**Example :**   
numbers = [1, 2, 3, 4, 5]  
squared = map(lambda x: x ** 2, numbers)  
print(list(squared))  # Output: [1, 4, 9, 16, 25]

**filter() Function :** Purpose: Filters elements in an iterable based on a condition defined by a function and returns an iterator containing only the elements that satisfy the condition.Selects a subset of the input iterable.The function must return True or False.

Example :    
numbers = [1, 2, 3, 4, 5]   
even_numbers = filter(lambda x: x % 2 == 0, numbers)   
print(list(even_numbers))  # Output: [2, 4]

**reduce() Function :** Purpose: Combines the elements of an iterable into a single value by repeatedly applying a function. It is part of the functools module.Performs cumulative operations, such as summing or multiplying elements.Returns a single value, not an iterable.

Example:   
from functools import reduce  
numbers = [1, 2, 3, 4, 5]   
product = reduce(lambda x, y: x * y, numbers)   
print(product)  # Output: 120

**1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list ?**

```python
def sum_of_evens(numbers):
  return sum(num for num in numbers if num % 2 == 0)

user_input = input("Enter numbers separated by spaces: ")
numbers = list(map(int, user_input.split()))
result = sum_of_evens(numbers)
print(f"The sum of even numbers is: {result}")
```

**4.Write a Python function that checks if a given number is prime or not from 1 to 200 ?**

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms ?**

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total terms to generate
        self.current = 0        # Current term count
        self.a, self.b = 0, 1   # Starting values for Fibonacci sequence

    def __iter__(self):
        return self  # The class itself is the iterator

    def __next__(self):
        if self.current >= self.n_terms:
            raise StopIteration  # End the iteration
        if self.current == 0:
            self.current += 1
            return self.a  # Return the first Fibonacci number
        elif self.current == 1:
            self.current += 1
            return self.b  # Return the second Fibonacci number
        else:
            # Calculate the next Fibonacci number
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value  # Update values
            self.current += 1
            return next_value

# Example usage
n = int(input("Enter the number of terms for the Fibonacci sequence: "))
fib_iterator = FibonacciIterator(n)

print("Fibonacci sequence:")
for number in fib_iterator:
    print(number, end=" ")
```

**6.Write a generator function in Python that yields the powers of 2 up to a given exponent ?**

```python
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

max_exp = int(input("Enter the maximum exponent: "))
print("Powers of 2:")
for power in powers_of_two(max_exp):
    print(power)
```

**7.Implement a generator function that reads a file line by line and yields each line as a string ?**



```python
def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")

file_path = input("Enter the file path: ")
print("File contents line by line:")
for line in read_file(file_path):
    print(line)
```

**9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit ?**

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = list(map(float, input("Enter temperatures in Celsius separated by spaces: ").split()))
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string ?**

```python
def is_not_vowel(char):
    return char.lower() not in 'aeiou'
user_input = input("Enter a string: ")
filtered_string = ''.join(filter(is_not_vowel, user_input))
print("String after removing vowels:", filtered_string)
```

**11.Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 € ?**

**Write a Python program using lambda and map**

```python
# Sample data: List of orders where each order is represented as [order_number, price_per_item, quantity]
orders = [
    [34587, 40.95, 4],
    [98762, 56.80, 5],
    [77226, 32.95, 3],
    [88112, 24.99, 3],
    [10, 110.00, 100],
    [11,99.00,100]
]

# Using map and lambda to calculate the required tuples
order_totals = list(map(
    lambda order: (order[0], order[1] * order[2] + 10) if order[1] * order[2] < 10000 else (order[0], order[1] * order[2]),
    orders
))

# Displaying the results
print("Order totals:")
for order in order_totals:
    print(f"Order {order[0]}: €{order[1]:.2f}")
```
